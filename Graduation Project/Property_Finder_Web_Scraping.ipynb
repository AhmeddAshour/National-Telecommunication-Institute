```python
import requests

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

base_url = 'https://www.propertyfinder.eg/ar/search?l=2254-30754-24-20663-17-30504-5-20-19-2-13-10&c=1&fu=0&ob=mr&page='

with requests.Session() as s:
    responses = []
    for i in range(1, 501):  # Change range(1, 501) for the first 500 pages
        url = base_url + str(i)
        response = s.get(url, headers=headers)
        responses.append(response)

# Now 'responses' contains the response objects for each URL
```

This code will send requests for the first 500 pages (you can adjust the range as needed) and store the response objects in a list called responses.

##Processing Data

```python
import re
import json

# Assuming responses contain the HTTP responses from the first 10 URLs
for i, response in enumerate(responses):
    match = re.search(r'!--(\{"queryState".*?)-->', response.text)
    print(f'Match for URL {i+1}:', match)  # Print the match object to see if it's None

    if match:
        data = json.loads(match.group(1))
        print(f'Data from URL {i+1}:', data)  # Optionally print the parsed JSON data
    else:
        print(f"No match found in response from URL {i+1}.")
        # Handle the case where no match is found, e.g., by inspecting response.text
```

This code iterates through the collected responses, searches for the specified pattern in each response text, and processes the data if a match is found.

```python
# Print a snippet of the response text for each URL
for i, response in enumerate(responses):
    print(f'Response text from URL {i+1} (first 1000 characters):')
    print(response.text[:1000])  # Print the first 1000 characters for brevity
    print('-' * 80)  # Separator for clarity
```

This code prints the first 1000 characters of each response's text to give you a glimpse of the content without overwhelming output. You can adjust the number of characters as needed.

```python
import re
import json

data_list = []  # To store data for each response

for i, response in enumerate(responses):
    match = re.search(r'<script id="__NEXT_DATA__" type="application/json">({.*?})</script>', response.text)
    if match:
        data = json.loads(match.group(1))
        data_list.append(data)
        print(f'Data from URL {i+1} extracted successfully.')
    else:
        print(f"No match found in response from URL {i+1}.")

# data_list now contains the extracted data for each URL
```

This code iterates over each response, searches for the specified script tag, and stores the parsed JSON data in a list called data_list.

##Making DataFrame

```python
import warnings
import pandas as pd
from pandas import json_normalize

warnings.filterwarnings('ignore')

# Sample datalist with JSON data extracted from the responses
datalist = [json.loads(re.search(r'<script id="__NEXT_DATA__" type="application/json">({.*?})</script>', response.text).group(1)) for response in responses]

def make_dataframe(datalist):
    all_listings = []  # List to store all listings from each page

    for data in datalist:
        # Adjust path to reach 'listings' within the JSON structure
        listings = data.get('props', {}).get('pageProps', {}).get('searchResult', {}).get('listings', [])
        # Add listings to the all_listings list
        all_listings.extend(listings)

    # Use json_normalize to flatten the nested dictionary structure
    return json_normalize(all_listings)

# Create the DataFrame with the flattened listings data
df = make_dataframe(datalist)

# Display the first few rows of the DataFrame to check
df.head()
```

In this code, datalist is populated with JSON data directly extracted from the responses using a list comprehension. The rest of the code processes this data into a pandas DataFrame.

```python
import pandas as pd

# Function to drop columns with the same value in all rows
def drop_constant_columns(df):
    # Iterate over each column and drop it if all values are the same
    for column in df.columns:
        # Check if the column contains lists
        if df[column].apply(lambda x: isinstance(x, list)).any():
            # Handle list columns (e.g., skip or apply a different logic)
            print(f"Column '{column}' contains lists and will be skipped.")
            continue  # Skip to the next column

        if df[column].nunique() == 1:
            df.drop(column, axis=1, inplace=True)
    return df

# Assuming df is your DataFrame
# Make sure to define and populate your DataFrame before calling the function.
# Example:
# df = pd.DataFrame({'col1': [1, 1, 1], 'col2': ['a', 'b', 'c'], 'col3': [[1, 2], [1, 2], [1, 2]]})
df = drop_constant_columns(df)

# Print the DataFrame after dropping constant columns
print(df)
```

This function checks each column in the DataFrame and removes those with a constant value across all rows, ensuring that only columns with varying data are retained.
